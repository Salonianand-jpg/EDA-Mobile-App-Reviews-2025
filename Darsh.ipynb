# Data Cleaning

```python
# Fill missing categorical values with 'Unknown'
df['user_country'] = df['user_country'].fillna('Unknown')
df['user_gender'] = df['user_gender'].fillna('Unknown')

# Fill missing numeric values in 'rating' with mean
df['rating'] = df['rating'].fillna(df['rating'].mean())

# Check categorical and numeric columns
print("Missing values after filling:")
print(df[['user_country', 'user_gender', 'rating']].isnull().sum())  # Should print 0

# Remove rows with missing values in important columns
df.dropna(subset=['app_version', 'review_text'], inplace=True)

# Convert user_age to integer
df['user_age'] = df['user_age'].astype(int)

# Convert review_date to datetime
df['review_date'] = pd.to_datetime(df['review_date'], errors='coerce')

# Check if any missing values remain in the dataset
print("Rows with any missing values:")
print(df[df.isnull().any(axis=1)])  # Should be empty if all missing rows removed

# Show total number of rows and columns after cleaning
total_rows, total_columns = df.shape
print(f"Total number of rows after cleaning: {total_rows}")
print(f"Total number of columns after cleaning: {total_columns}")

# Check updated data types
print(df.dtypes)

# Check if any invalid dates became NaT after conversion
print("Invalid dates converted to NaT:", df['review_date'].isna().sum())
```

#Removing missing values with another value and dropping rows that contains missing values:

Removed missing values with 'unknown' for columns user_country,user_gender,rating using 'fillna()'. Dropped rows that contains missing values that are not necessary for analysis using 'dropna()'. And after removing and dropping, checks the shape of the cleaned dataset using 'df.shape'.

#Changing DataType:

Changed the DataType of columns user_age using 'astype()' and review_date using 'to_datetime()'. After conversion, checks the updated data types using 'df.dtypes' and also checks if any invalid dates became NaT after conversion.



# Adding New Columns

```python
# Unique identifier based on user + app + review date
df['unique_review_id'] = df['user_id'].astype(str) + '_' + df['app_name'] + '_' + df['review_date'].astype(str)

# Check duplicates based on this new column
duplicate_reviews = df[df['unique_review_id'].duplicated(keep=False)]
print("Number of duplicate reviews:", duplicate_reviews.shape[0])
```

```python
# Add a new column 'ratings_group' based on rating

df['ratings_category']=np.where(df['rating']>=5, 'High', 
                             np.where(df['rating']<4, 'Medium','Low'))

# Add a new column 'age_group' based on user_age

df['age_group']= pd.cut(df['user_age'], bins=[0,18,30,45,60,100], 
                        labels=['Teen','Young Adult','Adult','Middle Age','Senior'])

#  Add a new column 'review_year' and 'review_month' based on review_date

df['review_year'] = df['review_date'].dt.year
df['review_month'] = df['review_date'].dt.month
df
```

# Data Visualization

**Problem Statement: Top 10 Apps by Number of Reviews.**

```python
import matplotlib.pyplot as plt
import pandas as pd

# Count number of reviews per app
top_apps = df['app_name'].value_counts().head(10)

# Plot bar chart
plt.figure(figsize=(10,5))
bars = plt.bar(top_apps.index, top_apps.values, color='purple')

# Add exact values on top of bars 
plt.bar_label(bars, labels=top_apps.values, fontsize=10, fontweight='bold')

# Add title and labels
plt.title('Top 10 Apps by Number of Reviews', fontsize=16)
plt.xlabel('App Name', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Inference:
- **Pinterest** has the **highest** number of reviews **77**, indicating high user engagement and popularity.
- Close competitors are **Reddit** with **75** and **OneDrive** with **74**.
- At **bottom** **Grammarly** with **65** reviews, but **still competitive** since the gap is small.
- The **difference between the top and bottom app is small**, which implies user engagement is **balanced**.
